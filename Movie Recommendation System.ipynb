Importing the dependencies

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer #To enumerate data into data so that machine can understand
from sklearn.metrics.pairwise import cosine_similarity #To find the similarity between movies
```

Data Collection and Pre-Processing

```python
#loading the data from the csv file to a pandas dataframe
movies = pd.read_csv('movies.csv')

#printing the first five rows of the data frame
movies.head()
```

```python
movies
```

```python
#No of rows and columns
movies.shape
movies.columns
```

```python
movies.describe()
```

```python
#Selecting the relevant features for recommendation
movies = movies[['id','title','overview','genre']]
```

```python
movies['tags'] = movies['overview'] + movies['genre']
```

```python
new_data = movies.drop(columns=['overview','genre'])
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')  #we will get output in only english language movies and total features(here,movies) are 10000
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
def recommend(movies):
  index = new_data[new_data['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```

```python
pickle.dump(new_data,open('movies.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.load(open('movies.pkl','rb'))
```
